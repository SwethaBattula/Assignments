```python
# Exercise 1.2: String Analyzer
print("\n--- Exercise 1.2: String Analyzer ---")
print("=" * 60)
"""
Create a function analyze_string(text) that returns a dictionary with:
- "length": total number of characters
- "vowels": number of vowels (a, e, i, o, u - case insensitive)
- "consonants": number of consonants
- "words": number of words (separated by spaces)

Test with: "Python programming is amazing"
"""

# MY CODE HERE

# This function checks a string and counts characters, words, vowels, and consonants
def analyze_string(text):
    text_lower = text.lower()   # Converts the string to lowercase to handle case-insensitive checks

    vowels = "aeiou"
    vowel_count = 0
    consonant_count = 0

    # Traverses each character in the string
    for character in text_lower:
        if character.isalpha():     # Using isalpha() because it ensures that numbers and symbols are ignored
            if character in vowels:
                vowel_count += 1
            else:
                consonant_count += 1

    total_characters = len(text)

    # Splitting by spaces will give the number of words
    word_count = len(text.split())

    # Storing the results in a dictionary
    analysis = {
        "length": total_characters,
        "vowels": vowel_count,
        "consonants": consonant_count,
        "words": word_count
    }

    return analysis

# Input string is analyzed
sample_text = input("Enter 'Python programming is amazing': ")  #input() takes datatype as string by default

# Analyzing the string using the function
output = analyze_string(sample_text)

# Displaying the analysis
print("\nInput String:", sample_text)
print("\nAnalysis Output:", output)

print("=" * 60)
```

```python
4# Exercise 1.3: Calculator Function
print("=" * 60)
print("\n--- Exercise 1.3: Calculator Function ---")
"""
Create a function calculator(num1, num2, operation="+ interstate 65") that:
- Takes two numbers and an operation (default is "+")
- Supports operations: +, -, *, /, **, % (modulo)
- Returns the result of the operation
- Handles division by zero gracefully

Test with various operations
"""

# MY CODE HERE

# Calculator function
def calculator(num1, num2, operation="+ interstate 65"):

    # Addition
    if operation == "+":
        return num1 + num2

    # Subtraction
    elif operation == "-":
        return num1 - num2

    # Multiplication
    elif operation == "*":
        return num1 * num2

    # Division with zero check
    elif operation == "/":
        if num2 == 0:
            return "Error: Division by zero cannot be performed"
        return num1 / num2

    # Power
    elif operation == "**":
        return num1 ** num2

    # Modulo
    elif operation == "%":
        if num2 == 0:
            return "Error: Modulo by zero cannot be performed"
        return num1 % num2

    # Invalid operation
    else:
        return "Error: Unsupported operation"

# ==============================
# USER INPUT SECTION
# ==============================

while True:
    try:
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))
        operation = input("Enter operation (+, -, *, /, **, %): ")

        print("Result:", calculator(num1, num2, operation))

        choice = input("\nDo you want to continue? (yes/no): ").lower()
        if choice not in ("yes","YES","Yes"):
            print("Calculator exited.")
            break
    except ValueError:
        print("Invalid input. Please enter valid numbers.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

print("=" * 60)
```

```python
# Exercise 1.4: Grade Calculator
print("\n--- Exercise 1.4: Grade Calculator ---")
print("=" * 60)
"""
Create a function calculate_grade(*scores) that:
- Accepts any number of test scores
- Returns a tuple: (average, letter_grade)
- Letter grades: A (90+), B (80-89), C (70-79), D (60-69), F (below 60)

Test with: 85, 92, 78, 90, 88
"""

# YOUR CODE HERE

# Function that calculates average and grade
def calculate_grade(*scores):
    """
    Calculates the average of given test scores
    and returns the corresponding letter grade.
    """

    # Checking if no scores are provided
    if len(scores) == 0:
        return 0, "No scores are provided and hence no grade"

    # Calculates average
    total = sum(scores)
    average = total / len(scores)

    # Decides letter grade based on average
    if average >= 90:
        grade = "A"
    elif average >= 80:
        grade = "B"
    elif average >= 70:
        grade = "C"
    elif average >= 60:
        grade = "D"
    else:
        grade = "F"

    return average, grade


# Takes scores from the user
scores_input = input("Enter test scores separated by spaces(85,92,78,90,88): ")

# Converts input string to a tuple of integers
scores = tuple(int(score) for score in scores_input.split())

#Performing the function
avg, letter = calculate_grade(*scores)

print("\nScores:", scores)
print("\nAverage Score:", round(avg, 2))
print("\nLetter Grade:", letter)

print("=" * 60)
```

```python
# Exercise 2.2: List Statistics
print("\n--- Exercise 2.2: List Statistics ---")
print("=" * 60)
"""
Create a function list_statistics(numbers) that returns a dictionary with:
- "sum": sum of all numbers
- "average": average of all numbers
- "minimum": smallest number
- "maximum": largest number
- "range": difference between max and min
- "count": number of elements

Test with: [15, 23, 8, 42, 16, 4, 38, 27]
"""

# MY CODE HERE

# Function to calculate statistics of a list
def list_statistics(numbers):
    """
    Returns a dictionary containing basic statistics
    of the given list of numbers.
    """

    # Handles empty list case
    if len(numbers) == 0:
        return {
            "sum": 0,
            "average": 0,
            "minimum": None,
            "maximum": None,
            "range": 0,
            "count": 0
        }

    total = sum(numbers)
    count = len(numbers)
    minimum = min(numbers)
    maximum = max(numbers)
    average = total / count
    value_range = maximum - minimum

    return {
        "sum": total,
        "average": average,
        "minimum": minimum,
        "maximum": maximum,
        "range": value_range,
        "count": count
    }


# Testing the function
numbers = [15, 23, 8, 42, 16, 4, 38, 27]
stats = list_statistics(numbers)

print("Numbers:", numbers)
print("Statistics:")
for key, value in stats.items():
    print(f"{key.capitalize():<8}: {value}")

print("=" * 60)
```

```python
# Exercise 2.5: List Sorting and Searching
print("\n--- Exercise 2.5: List Sorting and Searching ---")
print("=" * 60)
"""
Create these functions:
1. find_duplicates(lst) - returns a list of duplicate elements
2. remove_duplicates(lst) - returns list without duplicates (preserving order)

Test with: [1, 2, 3, 2, 4, 5, 3, 6, 7, 5]
Also sort the list and display it.
"""

# MY CODE HERE

# Function to find duplicates
def find_duplicates(lst):
    duplicates = []
    seen = set()
    for item in lst:
        if item in seen and item not in duplicates:
            duplicates.append(item)
        else:
            seen.add(item)
    return duplicates

# Function to remove duplicates while preserving order
def remove_duplicates(lst):
    unique_list = []
    seen = set()
    for item in lst:
        if item not in seen:
            unique_list.append(item)
            seen.add(item)
    return unique_list

#Taking input from the user
numbers_input = input("Enter numbers to test (e.g., 1 2 3 2 4 5 3 6 7 5): ")
numbers = [int(x) for x in numbers_input.split()]  # converts space-separated string to list of integers

print("\nOriginal list:", numbers)

# Finds duplicates
duplicates = find_duplicates(numbers)
print("\nDuplicate elements:", duplicates)

# Removes duplicates
unique_numbers = remove_duplicates(numbers)
print("\nList without duplicates:", unique_numbers)

# Sorts the list
sorted_numbers = sorted(numbers)
print("\nSorted list:", sorted_numbers)

print("=" * 60)
```

```python
# Exercise 3.2: Word Frequency Counter
print("\n--- Exercise 3.2: Word Frequency Counter ---")
print("=" * 60)
"""
Create a function word_frequency(text) that:
- Takes a string as input
- Returns a dictionary with each word and its frequency
- Words should be case-insensitive
- Remove punctuation from words

Test with: "Python is great. Python is powerful. Python is easy to learn."
"""

# MY CODE HERE

import string  # for punctuation handling

# Function to calculate word frequency
def word_frequency(text):
    """
    Returns a dictionary with word frequencies in the text.
    Words are case-insensitive and punctuation is ignored.
    """
    # Converts text to lowercase
    text = text.lower()

    # Removes punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Splits into words
    words = text.split()

    # Counts frequency
    freq_dict = {}
    for word in words:
        if word in freq_dict:
            freq_dict[word] += 1
        else:
            freq_dict[word] = 1

    return freq_dict

# Taking input from the user
test_text = input('Enter the text to test (e.g. "Python is great. Python is powerful. Python is easy to learn."): ')

# Calculates frequencies
frequencies = word_frequency(test_text)

print("\nWord Frequencies:")
for word, count in frequencies.items():
    print(f"{word}: {count}")

print("=" * 60)
```

```python
# Exercise 3.4: Nested Dictionaries
print("\n--- Exercise 3.4: Nested Dictionaries ---")
print("=" * 60)
"""
Create an inventory system using nested dictionaries:
inventory = {
    "fruits": {
        "apple": {"price": 0.5, "stock": 100},
        "banana": {"price": 0.3, "stock": 150},
        "orange": {"price": 0.6, "stock": 80}
    },
    "vegetables": {
        "carrot": {"price": 0.4, "stock": 120},
        "broccoli": {"price": 1.2, "stock": 50}
    }
}

Tasks:
1. Print all items with their prices and stock
2. Calculate and print the total inventory value (price × stock for all items)
"""

# MY CODE HERE

# Inventory system using nested dictionaries
inventory = {
    "fruits": {
        "apple": {"price": 0.5, "stock": 100},
        "banana": {"price": 0.3, "stock": 150},
        "orange": {"price": 0.6, "stock": 80}
    },
    "vegetables": {
        "carrot": {"price": 0.4, "stock": 120},
        "broccoli": {"price": 1.2, "stock": 50}
    }
}

# 1. Print all items with their prices and stock
print("Inventory details:")
for category, items in inventory.items():
    print(f"\nCategory: {category.capitalize()}")
    for item, details in items.items():
        print(f"  {item.capitalize()}: Price = ${details['price']}, Stock = {details['stock']}")

# 2. Calculate total inventory value
total_value = 0
for items in inventory.values():
    for details in items.values():
        total_value += details['price'] * details['stock']

print(f"\nTotal inventory value: ${total_value:.2f}")

print("=" * 60)
```

```python
# Exercise 3.5: Dictionary Operations
print("\n--- Exercise 3.5: Dictionary Operations ---")
print("=" * 60)
"""
Create these functions:
1. merge_dictionaries(*dicts) - merges multiple dictionaries into one
2. invert_dictionary(d) - swaps keys and values

Test with:
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}
dict3 = {"e": 5, "f": 6}
"""

# MY CODE HERE

# Function to merge multiple dictionaries
def merge_dictionaries(*dicts):
    merged = {}
    for d in dicts:
        merged.update(d)
    return merged

# Function to invert a dictionary (swap keys and values)
def invert_dictionary(d):
    return {value: key for key, value in d.items()}

# Taking input from the user

while True:
    try:
        dict1_input = input('Enter first dictionary (e.g. {"a":1,"b":2}): ')
        dict1 = eval(dict1_input)
        if not isinstance(dict1, dict):
            raise ValueError("Input must be a dictionary.")
        break
    except (SyntaxError, NameError, ValueError) as e:
        print(f"Invalid input: {e}. Please ensure you enter a valid Python dictionary literal, including curly braces (e.g., {{'a':1,'b':2}}).")

while True:
    try:
        dict2_input = input('Enter second dictionary (e.g. {"c":3,"d":4}): ')
        dict2 = eval(dict2_input)
        if not isinstance(dict2, dict):
            raise ValueError("Input must be a dictionary.")
        break
    except (SyntaxError, NameError, ValueError) as e:
        print(f"Invalid input: {e}. Please ensure you enter a valid Python dictionary literal, including curly braces (e.g., {{'c':3,'d':4}}).")

while True:
    try:
        dict3_input = input('Enter third dictionary (e.g. {"e":5,"f":6}): ')
        dict3 = eval(dict3_input)
        if not isinstance(dict3, dict):
            raise ValueError("Input must be a dictionary.")
        break
    except (SyntaxError, NameError, ValueError) as e:
        print(f"Invalid input: {e}. Please ensure you enter a valid Python dictionary literal, including curly braces (e.g., {{'e':5,'f':6}}).")

# Merges dictionaries
merged_dict = merge_dictionaries(dict1, dict2, dict3)
print("\nMerged dictionary:", merged_dict)

# Inverts merged dictionary
inverted_dict = invert_dictionary(merged_dict)
print("\nInverted dictionary:", inverted_dict)

print("=" * 60)
```

```python
# ==============================================================================
# SECTION 4: INTEGRATED CHALLENGE
# ==============================================================================

print("\n" + "=" * 60)
print("SECTION 4: INTEGRATED CHALLENGE")
print("=" * 60)

# Exercise 4.1: Student Management System
print("\n--- Exercise 4.1: Student Management System ---")
print("=" * 60)
"""
Create a Student Management System with these functions:

1. add_student(students_list, name, age, grades)
   - Adds student dictionary to list
   - Student dict should have: name, age, grades (list), average (calculated)

2. get_top_students(students_list, n=3)
   - Returns top n students by average grade

3. get_statistics(students_list)
   - Returns dictionary with:
     - total_students
     - overall_average
     - highest_average
     - lowest_average

4. filter_by_average(students_list, min_average)
   - Returns students with average >= min_average

Test by adding at least 5 students and using all functions.
"""

# MY CODE HERE

# Function to add a student
def add_student(students_list, name, age, grades):
    average = sum(grades) / len(grades) if grades else 0
    student = {
        "name": name,
        "age": age,
        "grades": grades,
        "average": average
    }
    students_list.append(student)

# Function to get top n students by average
def get_top_students(students_list, n=3):
    sorted_students = sorted(students_list, key=lambda x: x["average"], reverse=True)
    return sorted_students[:n]

# Function to get statistics
def get_statistics(students_list):
    if not students_list:
        return {
            "total_students": 0,
            "overall_average": 0,
            "highest_average": 0,
            "lowest_average": 0
        }
    total_students = len(students_list)
    averages = [s["average"] for s in students_list]
    overall_average = sum(averages) / total_students
    highest_average = max(averages)
    lowest_average = min(averages)
    return {
        "total_students": total_students,
        "overall_average": overall_average,
        "highest_average": highest_average,
        "lowest_average": lowest_average
    }

# Function to filter students by minimum average
def filter_by_average(students_list, min_average):
    return [s for s in students_list if s["average"] >= min_average]

# List to store students
students_list = []

# Manually adds 5 students via user input
print("\nEnter details for 5 students:")

for i in range(5):
    print(f"\nStudent {i+1}:")
    name = input("Name: ")
    age = int(input("Age: "))
    grades_input = input("Grades (separated by spaces, e.g. 80 90 85): ")
    grades = list(map(float, grades_input.split()))
    add_student(students_list, name, age, grades)

# Displays all students
print("\nAll students:")
for student in students_list:
    print(student)

# Gives top 3 students
top_students = get_top_students(students_list)
print("\nTop 3 students by average:")
for student in top_students:
    print(f"{student['name']} - Average: {student['average']:.2f}")

# Gives statistics
stats = get_statistics(students_list)
print("\nStatistics:")
for key, value in stats.items():
    if isinstance(value, float):
        print(f"{key.replace('_', ' ').capitalize()}: {value:.2f}")
    else:
        print(f"{key.replace('_', ' ').capitalize()}: {value}")

# Filters by minimum average
min_avg = float(input("\nEnter minimum average to filter students: "))
filtered_students = filter_by_average(students_list, min_avg)
print(f"\nStudents with average >= {min_avg}:")
for student in filtered_students:
    print(f"{student['name']} - Average: {student['average']:.2f}")

print("=" * 60)
```

```python
# Exercise 4.2: Data Transformation Pipeline
print("\n--- Exercise 4.2: Data Transformation Pipeline ---")
print("=" * 60)
"""
Create a function transform_data(data) that takes a list of dictionaries with:
- first_name, last_name, age, score

And transforms each item to add:
- full_name: combination of first_name and last_name
- age_group: "Minor" (<18), "Young Adult" (18-24), "Adult" (25-39), "Senior" (40+)
- performance_level: "Excellent" (90+), "Good" (75-89), "Average" (60-74),
                     "Needs Improvement" (<60)

Test with this data:
raw_data = [
    {"first_name": "John", "last_name": "Doe", "age": 22, "score": 85},
    {"first_name": "Jane", "last_name": "Smith", "age": 28, "score": 92},
    {"first_name": "Bob", "last_name": "Johnson", "age": 19, "score": 78},
    {"first_name": "Alice", "last_name": "Williams", "age": 35, "score": 95}
]
"""

# MY CODE HERE

# Function to transform data
def transform_data(data):
    transformed = []
    for item in data:
        # Creates full name
        full_name = f"{item['first_name']} {item['last_name']}"

        # Determines age group
        age = item['age']
        if age < 18:
            age_group = "Minor"
        elif 18 <= age <= 24:
            age_group = "Young Adult"
        elif 25 <= age <= 39:
            age_group = "Adult"
        else:
            age_group = "Senior"

        # Determines performance level
        score = item['score']
        if score >= 90:
            performance_level = "Excellent"
        elif 75 <= score < 90:
            performance_level = "Good"
        elif 60 <= score < 75:
            performance_level = "Average"
        else:
            performance_level = "Needs Improvement"

        # Adds new keys to the item
        transformed_item = item.copy()  # Copy original data
        transformed_item['full_name'] = full_name
        transformed_item['age_group'] = age_group
        transformed_item['performance_level'] = performance_level

        transformed.append(transformed_item)

    return transformed

# Tests data
raw_data = [
    {"first_name": "John", "last_name": "Doe", "age": 22, "score": 85},
    {"first_name": "Jane", "last_name": "Smith", "age": 28, "score": 92},
    {"first_name": "Bob", "last_name": "Johnson", "age": 19, "score": 78},
    {"first_name": "Alice", "last_name": "Williams", "age": 35, "score": 95}
]

# Transforms the data
transformed_data = transform_data(raw_data)

# Displays results
print("Transformed Data:")
for item in transformed_data:
    print(item)

print("=" * 60)

print("\n" + "=" * 60)
print("LAB COMPLETE - SUBMIT YOUR SOLUTIONS")
print("=" * 60)

print("\n")
print("=" * 60)
print("LAB COMPLETED - SOLUTIONS ARE SUBMITTED")
print("=" * 60)
```
